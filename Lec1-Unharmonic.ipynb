```python
# THIS IS A ANHARMONIC OSCILLATOR
# POTENTIAL ENERGY V(x) = x**2/2 + x**4/4
# FORCE F(x) = -x-x**3


import numpy as np
import scipy.special as spc
import scipy.optimize as opt
import scipy.integrate as integrate
import matplotlib.pyplot as plt
from scipy.integrate import quad
import cmath
```

```python
nsteps=200

dt=0.05;dt2=dt*dt


plt.title("Anharmonic oscillator")


x=np.zeros(nsteps)

v=np.zeros(nsteps)

x[0]=1;v[0]=0.0

E0=v[0]**2/2+x[0]**2/2+x[0]**4/4

ai=-x[0]-x[0]**3


for i in range(nsteps-1):

   x[i+1]=x[i]+v[i]*dt+0.5*ai*dt2

   vi_half=v[i]+0.5*ai*dt

   #ai=-x[i+1]-x[i+1]**3 # unharmonic
   ai=-x[i+1]-x[i+1]**5    # harmonic

   v[i+1]=vi_half+0.5*ai*dt



```

```python
# WE TEST THE CONSERVATION OF THE TOTAL MECHANICAL ENERGY

E=v**2/2+x**2/2+x**4/4

plt.plot(np.array(range(nsteps))*dt,E,'o-',label="Total Energy vs. time")

plt.legend()

plt.show()
```

```python
plt.plot(x,v,label="Phase space trajectory")

plt.legend()

plt.show()
```
